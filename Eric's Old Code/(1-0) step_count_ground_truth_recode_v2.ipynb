```python
import pandas as pd
import numpy as np
import openpyxl
import os
import re
from datetime import datetime, timedelta
```

```python
def parse_observation(observation):

    ob = re.split(r'[-_]', observation)

    id = int(ob[1])
    session = int(ob[2])
    return id, session


col = np.array(['ID', 'Session', 'Relative Time', 'Date', 'Time', 'Quality', 'Step'])
```

```python
directory = "C:/Users/HELIOS-300/Desktop/Data/Stepcount _10112024 - Step_116_02_MD - Event Logs.xlsx"
workbook = openpyxl.load_workbook(directory)
sheet = workbook[workbook.sheetnames[0]]
df = pd.DataFrame(sheet.values)
df.columns = np.array(df.loc[0])
df = df.drop(0)
df.head(10)
id, session = parse_observation(df.loc[1]['Observation'])

mask = (log_final['id'] == id) & (log_final['do'] == session)
observation = log_final[mask].iloc[0]
df



idxs = (log_final['id'] <= id) & (log_final['do'] == session)

log_final[idxs]

```

```python
directory = "C:/Users/HELIOS-300/Desktop/WAVES/Steps Data WAVES/steps"

files = os.listdir(directory)
files = [f for f in files if 'xlsx' in f]
step_ground_truth = []
for f in files:
    workbook = openpyxl.load_workbook(directory + '/' + f)
    sheet = workbook[workbook.sheetnames[0]]
    df = pd.DataFrame(sheet.values)
    df.columns = np.array(df.loc[0])
    df = df.drop(0)
    df.head(10)
    id, session = parse_observation(df.loc[1]['Observation'])

    mask = (log_final['id'] == id) & (log_final['do'] == session)
    observation = log_final[mask].iloc[0]


    start_time = observation['start_time']
    month = observation['start_month']
    day = observation['start_day']
    year = observation['start_year']
    date = str(month) + '/' + str(day) + '/' + str(year)

    quality = ''

    for idx, row in df.iterrows():
        if row['Behavior'] == 'step':
            rel_time = row['Time_Relative_hmsf']
            
            start_format = "%H:%M:%S %p"    
            rel_format = "%H:%M:%S.%f"
            start = datetime.strptime(start_time, start_format)

            result_time = start + rel_time

            time = timedelta(hours=result_time.hour, minutes=result_time.minute, seconds=result_time.second, microseconds=result_time.microsecond)
            step_row = [id, session, rel_time, date, time, quality, 1]
  
            step_ground_truth.append(step_row)
        else:
            quality = row['Behavior']

step_ground_truth_recode = pd.DataFrame(data=np.array(step_ground_truth), columns=col)
step_ground_truth_recode.loc[
    (step_ground_truth_recode['Quality'] == 'Non-codeable') & 
    (step_ground_truth_recode['Step'] >= 1), 
    'Quality'
] = 'Codable'

step_ground_truth_recode
```

```python
directory = "C:/Users/HELIOS-300/Desktop/WAVES/Steps Data WAVES/steps"
files = os.listdir(directory)
files = [f for f in files if 'xlsx' in f]
seconds_ground_truth = []
for f in files:
    workbook = openpyxl.load_workbook(directory + '/' + f)
    sheet = workbook[workbook.sheetnames[0]]
    df = pd.DataFrame(sheet.values)
    df.columns = np.array(df.loc[0])
    df = df.drop(0)
    df.head(10)
    id, session = parse_observation(df.loc[1]['Observation'])

    mask = (log_final['id'] == id) & (log_final['do'] == session)
    observation = log_final[mask].iloc[0]


    start_time = observation['start_time']
    month = observation['start_month']
    day = observation['start_day']
    year = observation['start_year']
    date = str(month) + '/' + str(day) + '/' + str(year)

    quality = ''
    step_counter = 0
    second = 1
    row_idx = 0



    while row_idx < len(df):
        if df.iloc[row_idx]['Behavior'] == 'step':
            if df.iloc[row_idx]['Time_Relative_hms'].seconds == second:
                step_counter += 1
                row_idx += 1
            else:
                start_format = "%H:%M:%S %p"    
                rel_format = "%H:%M:%S"
                rel_time = timedelta(seconds=second)
                
                
                start = datetime.strptime(start_time, start_format)

                # Add the times
                result_time = start + rel_time

                # Convert the result back to a formatted string
                time = result_time.strftime(rel_format)
                rel_time = timedelta(hours=rel_time.seconds // 3600,
                                minutes=(rel_time.seconds // 60) % 60,
                                seconds=rel_time.seconds % 60)
                
                seconds_row = [id, session, str(rel_time), date, time, quality, step_counter]
                # seconds_ground_truth.loc[len(seconds_ground_truth)] = seconds_row
                seconds_ground_truth.append(seconds_row)
                second += 1
                step_counter = 0

        else:
            quality = df.iloc[row_idx]['Behavior']
            row_idx += 1

    if step_counter > 0:
        start_format = "%H:%M:%S %p"    
        rel_format = "%H:%M:%S"
        rel_time = timedelta(seconds=second)
        
        
        start = datetime.strptime(start_time, start_format)

        result_time = start + rel_time

        time = result_time.strftime(rel_format)
        rel_time = timedelta(hours=rel_time.seconds // 3600,
                        minutes=(rel_time.seconds // 60) % 60,
                        seconds=rel_time.seconds % 60)
        
        seconds_row = [id, session, rel_time, date, time, quality, step_counter]
        seconds_ground_truth.append(seconds_row)

# Convert to DataFrame
seconds_ground_truth_recode = pd.DataFrame(data=np.array(seconds_ground_truth), columns=col)
seconds_ground_truth_recode.loc[
    (seconds_ground_truth_recode['Quality'] == 'Non-codeable') & 
    (seconds_ground_truth_recode['Step'] >= 1), 
    'Quality'
] = 'Codable'

# Show result
seconds_ground_truth_recode
```
