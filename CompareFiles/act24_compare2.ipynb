## Step 2: Visualization Setup and Functions

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 6)
plt.rcParams['font.size'] = 10

print("Visualization libraries imported successfully!")
print("\nPackages being used:")
print("- matplotlib: For creating pie charts and bar charts")
print("- seaborn: For enhanced styling and color palettes")
print("- pandas: For data manipulation (already imported)")
```

```python
# Helper function to create comparison visualizations
def compare_categorical_columns(cam_col, gt_col, cam_df, gt_df, 
                                max_categories_for_pie=8, figsize=(16, 6)):
    """
    Compare categorical columns from two dataframes.
    Uses pie charts for small categories, bar charts for many categories.
    """
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # Get value counts for both columns
    cam_counts = cam_df[cam_col].value_counts()
    gt_counts = gt_df[gt_col].value_counts()
    
    # Determine if we should use pie or bar chart
    use_pie = len(cam_counts) <= max_categories_for_pie and len(gt_counts) <= max_categories_for_pie
    
    if use_pie:
        # Pie charts
        # cam_df pie chart
        axes[0].pie(cam_counts.values, labels=cam_counts.index, autopct='%1.1f%%', startangle=90)
        axes[0].set_title(f'cam_df: {cam_col}\n(n={len(cam_df)} rows)', fontsize=12, fontweight='bold')
        
        # gt_df pie chart
        axes[1].pie(gt_counts.values, labels=gt_counts.index, autopct='%1.1f%%', startangle=90)
        axes[1].set_title(f'gt_df: {gt_col}\n(n={len(gt_df)} rows)', fontsize=12, fontweight='bold')
    else:
        # Bar charts for many categories
        # cam_df bar chart
        cam_counts.plot(kind='bar', ax=axes[0], color='steelblue')
        axes[0].set_title(f'cam_df: {cam_col}\n(n={len(cam_df)} rows)', fontsize=12, fontweight='bold')
        axes[0].set_xlabel('Category')
        axes[0].set_ylabel('Count')
        axes[0].tick_params(axis='x', rotation=45)
        plt.setp(axes[0].xaxis.get_majorticklabels(), ha='right')
        axes[0].grid(axis='y', alpha=0.3)
        
        # gt_df bar chart
        gt_counts.plot(kind='bar', ax=axes[1], color='coral')
        axes[1].set_title(f'gt_df: {gt_col}\n(n={len(gt_df)} rows)', fontsize=12, fontweight='bold')
        axes[1].set_xlabel('Category')
        axes[1].set_ylabel('Count')
        axes[1].tick_params(axis='x', rotation=45)
        plt.setp(axes[1].xaxis.get_majorticklabels(), ha='right')
        axes[1].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print summary statistics
    print(f"\n{'='*60}")
    print(f"SUMMARY: {cam_col} (cam_df) vs {gt_col} (gt_df)")
    print(f"{'='*60}")
    print(f"\ncam_df - {cam_col}:")
    print(f"  Total unique values: {cam_df[cam_col].nunique()}")
    print(f"  Missing values: {cam_df[cam_col].isnull().sum()}")
    print(f"  Top 5 values:")
    for val, count in cam_counts.head(5).items():
        pct = (count / len(cam_df)) * 100
        print(f"    {val}: {count:,} ({pct:.1f}%)")
    
    print(f"\ngt_df - {gt_col}:")
    print(f"  Total unique values: {gt_df[gt_col].nunique()}")
    print(f"  Missing values: {gt_df[gt_col].isnull().sum()}")
    print(f"  Top 5 values:")
    for val, count in gt_counts.head(5).items():
        pct = (count / len(gt_df)) * 100
        print(f"    {val}: {count:,} ({pct:.1f}%)")
    print()
```

## Analysis: ID and Observation Combinations

```python
# Find total ID and obs combinations for both dataframes
print("="*60)
print("ID AND OBSERVATION COMBINATIONS ANALYSIS")
print("="*60)

# Get unique (id, obs) combinations for each dataframe
cam_combinations = cam_df[['id', 'obs']].drop_duplicates()
gt_combinations = gt_df[['id', 'observation']].drop_duplicates()

# Rename 'observation' to 'obs' in gt_df for consistency
gt_combinations_renamed = gt_combinations.rename(columns={'observation': 'obs'})

print(f"\ncam_df:")
print(f"  Total unique (ID, Obs) combinations: {len(cam_combinations):,}")
print(f"  Total rows: {len(cam_df):,}")

print(f"\ngt_df:")
print(f"  Total unique (ID, Obs) combinations: {len(gt_combinations_renamed):,}")
print(f"  Total rows: {len(gt_df):,}")

# Find combinations in cam_df but not in gt_df
cam_combinations_set = set(zip(cam_combinations['id'], cam_combinations['obs']))
gt_combinations_set = set(zip(gt_combinations_renamed['id'], gt_combinations_renamed['obs']))

only_in_cam = cam_combinations_set - gt_combinations_set
only_in_gt = gt_combinations_set - cam_combinations_set
in_both = cam_combinations_set & gt_combinations_set

print(f"\n{'='*60}")
print("COMBINATION OVERLAP")
print(f"{'='*60}")
print(f"  Combinations in BOTH dataframes: {len(in_both):,}")
print(f"  Combinations ONLY in cam_df: {len(only_in_cam):,}")
print(f"  Combinations ONLY in gt_df: {len(only_in_gt):,}")

# Count rows for combinations that are in cam_df but not in gt_df
only_in_cam_list = list(only_in_cam)
if len(only_in_cam_list) > 0:
    # Create a mask for rows in cam_df that match these combinations
    cam_mask = cam_df.apply(lambda row: (row['id'], row['obs']) in only_in_cam, axis=1)
    rows_only_in_cam = cam_df[cam_mask]
    
    print(f"\n{'='*60}")
    print("ROWS FOR COMBINATIONS IN cam_df BUT NOT IN gt_df")
    print(f"{'='*60}")
    print(f"  Number of rows: {len(rows_only_in_cam):,}")
    print(f"  Percentage of total cam_df: {(len(rows_only_in_cam)/len(cam_df)*100):.2f}%")
    
    # Show breakdown by combination
    combination_row_counts = rows_only_in_cam.groupby(['id', 'obs']).size().reset_index(name='row_count')
    print(f"\n  Breakdown by (ID, Obs) combination:")
    print(f"    Total unique combinations: {len(combination_row_counts):,}")
    print(f"    Average rows per combination: {combination_row_counts['row_count'].mean():.1f}")
    print(f"    Min rows per combination: {combination_row_counts['row_count'].min()}")
    print(f"    Max rows per combination: {combination_row_counts['row_count'].max()}")
    
    print(f"\n  Top 10 (ID, Obs) combinations by row count:")
    top_combos = combination_row_counts.nlargest(10, 'row_count')
    for idx, row in top_combos.iterrows():
        print(f"    ID {row['id']}, Obs {row['obs']}: {row['row_count']:,} rows")
else:
    print(f"\n  No combinations found only in cam_df")
```

## Analysis: Time Coverage Comparison by (ID, Obs) Combinations

```python
# Analyze time coverage for each unique (id, obs) combination
print("="*60)
print("TIME COVERAGE ANALYSIS BY (ID, OBS) COMBINATIONS")
print("="*60)

# Helper function to convert rel_time to seconds for calculations
def time_to_seconds(time_str):
    """Convert HH:MM:SS format to total seconds"""
    if pd.isna(time_str):
        return np.nan
    try:
        parts = str(time_str).split(':')
        if len(parts) == 3:
            hours = int(parts[0])
            minutes = int(parts[1])
            seconds = int(parts[2])
            return hours * 3600 + minutes * 60 + seconds
        return np.nan
    except:
        return np.nan

# Calculate time coverage for cam_df
cam_df['rel_time_seconds'] = cam_df['rel_time'].apply(time_to_seconds)
cam_coverage = cam_df.groupby(['id', 'obs']).agg({
    'rel_time_seconds': ['min', 'max', 'count'],
    'rel_time': ['first', 'last']
}).reset_index()
cam_coverage.columns = ['id', 'obs', 'min_seconds', 'max_seconds', 'row_count', 'first_rel_time', 'last_rel_time']
cam_coverage['duration_seconds'] = cam_coverage['max_seconds'] - cam_coverage['min_seconds']
cam_coverage['duration_hours'] = cam_coverage['duration_seconds'] / 3600

# Calculate time coverage for gt_df
gt_df['rel_time_seconds'] = gt_df['relative_time'].apply(time_to_seconds)
gt_coverage = gt_df.groupby(['id', 'observation']).agg({
    'rel_time_seconds': ['min', 'max', 'count'],
    'relative_time': ['first', 'last']
}).reset_index()
gt_coverage.columns = ['id', 'obs', 'min_seconds', 'max_seconds', 'row_count', 'first_rel_time', 'last_rel_time']
gt_coverage['duration_seconds'] = gt_coverage['max_seconds'] - gt_coverage['min_seconds']
gt_coverage['duration_hours'] = gt_coverage['duration_seconds'] / 3600

print(f"\ncam_df:")
print(f"  Total unique (ID, Obs) combinations: {len(cam_coverage):,}")
print(f"  Average duration per combination: {cam_coverage['duration_hours'].mean():.2f} hours")
print(f"  Median duration per combination: {cam_coverage['duration_hours'].median():.2f} hours")
print(f"  Min duration: {cam_coverage['duration_hours'].min():.2f} hours")
print(f"  Max duration: {cam_coverage['duration_hours'].max():.2f} hours")
print(f"  Average row count per combination: {cam_coverage['row_count'].mean():.1f}")

print(f"\ngt_df:")
print(f"  Total unique (ID, Obs) combinations: {len(gt_coverage):,}")
print(f"  Average duration per combination: {gt_coverage['duration_hours'].mean():.2f} hours")
print(f"  Median duration per combination: {gt_coverage['duration_hours'].median():.2f} hours")
print(f"  Min duration: {gt_coverage['duration_hours'].min():.2f} hours")
print(f"  Max duration: {gt_coverage['duration_hours'].max():.2f} hours")
print(f"  Average row count per combination: {gt_coverage['row_count'].mean():.1f}")

# Merge to compare coverage for matching (id, obs) combinations
comparison = cam_coverage.merge(
    gt_coverage,
    on=['id', 'obs'],
    how='outer',
    suffixes=('_cam', '_gt')
)

# Calculate differences
comparison['duration_diff_hours'] = comparison['duration_hours_cam'] - comparison['duration_hours_gt']
comparison['row_count_diff'] = comparison['row_count_cam'] - comparison['row_count_gt']

# Find combinations in both datasets
in_both = comparison[comparison['duration_hours_cam'].notna() & comparison['duration_hours_gt'].notna()]

print(f"\n{'='*60}")
print("COVERAGE COMPARISON FOR MATCHING (ID, OBS) COMBINATIONS")
print(f"{'='*60}")
print(f"  Combinations in BOTH datasets: {len(in_both):,}")
if len(in_both) > 0:
    print(f"  Average duration difference (cam - gt): {in_both['duration_diff_hours'].mean():.2f} hours")
    print(f"  Median duration difference: {in_both['duration_diff_hours'].median():.2f} hours")
    print(f"  Average row count difference (cam - gt): {in_both['row_count_diff'].mean():.1f} rows")
    print(f"  Median row count difference: {in_both['row_count_diff'].median():.1f} rows")
    
    # Count how many have longer/shorter coverage
    longer_in_cam = (in_both['duration_diff_hours'] > 0).sum()
    longer_in_gt = (in_both['duration_diff_hours'] < 0).sum()
    same_duration = (in_both['duration_diff_hours'] == 0).sum()
    
    print(f"\n  Duration comparison:")
    print(f"    Longer in cam_df: {longer_in_cam:,} ({longer_in_cam/len(in_both)*100:.1f}%)")
    print(f"    Longer in gt_df: {longer_in_gt:,} ({longer_in_gt/len(in_both)*100:.1f}%)")
    print(f"    Same duration: {same_duration:,} ({same_duration/len(in_both)*100:.1f}%)")

# Create visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Plot 1: Distribution of durations - cam_df
axes[0, 0].hist(cam_coverage['duration_hours'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Duration (hours)', fontsize=11)
axes[0, 0].set_ylabel('Number of (ID, Obs) Combinations', fontsize=11)
axes[0, 0].set_title(f'cam_df: Duration Distribution\n(n={len(cam_coverage):,} combinations)', 
                    fontsize=12, fontweight='bold')
axes[0, 0].grid(axis='y', alpha=0.3)

# Plot 2: Distribution of durations - gt_df
axes[0, 1].hist(gt_coverage['duration_hours'], bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[0, 1].set_xlabel('Duration (hours)', fontsize=11)
axes[0, 1].set_ylabel('Number of (ID, Obs) Combinations', fontsize=11)
axes[0, 1].set_title(f'gt_df: Duration Distribution\n(n={len(gt_coverage):,} combinations)', 
                    fontsize=12, fontweight='bold')
axes[0, 1].grid(axis='y', alpha=0.3)

# Plot 3: Side-by-side comparison of durations
if len(in_both) > 0:
    axes[0, 2].scatter(in_both['duration_hours_gt'], in_both['duration_hours_cam'], 
                       alpha=0.5, s=20, color='purple')
    # Add diagonal line
    max_val = max(in_both['duration_hours_gt'].max(), in_both['duration_hours_cam'].max())
    axes[0, 2].plot([0, max_val], [0, max_val], 'r--', linewidth=2, label='Equal duration')
    axes[0, 2].set_xlabel('gt_df Duration (hours)', fontsize=11)
    axes[0, 2].set_ylabel('cam_df Duration (hours)', fontsize=11)
    axes[0, 2].set_title(f'Duration Comparison: cam_df vs gt_df\n(n={len(in_both):,} matching combinations)', 
                        fontsize=12, fontweight='bold')
    axes[0, 2].legend()
    axes[0, 2].grid(alpha=0.3)

# Plot 4: Distribution of duration differences
if len(in_both) > 0:
    axes[1, 0].hist(in_both['duration_diff_hours'], bins=50, color='green', edgecolor='black', alpha=0.7)
    axes[1, 0].axvline(x=0, color='red', linestyle='--', linewidth=2, label='No difference')
    axes[1, 0].set_xlabel('Duration Difference (cam_df - gt_df) in hours', fontsize=11)
    axes[1, 0].set_ylabel('Number of Combinations', fontsize=11)
    axes[1, 0].set_title('Distribution of Duration Differences', fontsize=12, fontweight='bold')
    axes[1, 0].legend()
    axes[1, 0].grid(axis='y', alpha=0.3)

# Plot 5: Row count comparison
if len(in_both) > 0:
    axes[1, 1].scatter(in_both['row_count_gt'], in_both['row_count_cam'], 
                      alpha=0.5, s=20, color='orange')
    max_rows = max(in_both['row_count_gt'].max(), in_both['row_count_cam'].max())
    axes[1, 1].plot([0, max_rows], [0, max_rows], 'r--', linewidth=2, label='Equal row count')
    axes[1, 1].set_xlabel('gt_df Row Count', fontsize=11)
    axes[1, 1].set_ylabel('cam_df Row Count', fontsize=11)
    axes[1, 1].set_title(f'Row Count Comparison\n(n={len(in_both):,} matching combinations)', 
                        fontsize=12, fontweight='bold')
    axes[1, 1].legend()
    axes[1, 1].grid(alpha=0.3)

# Plot 6: Top 20 combinations by duration difference
if len(in_both) > 0:
    top_diff = in_both.nlargest(20, 'duration_diff_hours')[['id', 'obs', 'duration_diff_hours']]
    top_diff['label'] = top_diff.apply(lambda x: f"ID {x['id']}, Obs {x['obs']}", axis=1)
    
    axes[1, 2].barh(range(len(top_diff)), top_diff['duration_diff_hours'].values, 
                   color='steelblue', alpha=0.7)
    axes[1, 2].set_yticks(range(len(top_diff)))
    axes[1, 2].set_yticklabels(top_diff['label'], fontsize=9)
    axes[1, 2].set_xlabel('Duration Difference (hours)', fontsize=11)
    axes[1, 2].set_title('Top 20 Combinations: Longer in cam_df', fontsize=12, fontweight='bold')
    axes[1, 2].axvline(x=0, color='red', linestyle='--', linewidth=1)
    axes[1, 2].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Print detailed statistics
print(f"\n{'='*60}")
print("DETAILED STATISTICS")
print(f"{'='*60}")

if len(in_both) > 0:
    print(f"\nTop 10 combinations with largest duration differences (cam_df longer):")
    top_cam = in_both.nlargest(10, 'duration_diff_hours')
    for idx, row in top_cam.iterrows():
        print(f"  ID {row['id']}, Obs {row['obs']}: "
              f"cam={row['duration_hours_cam']:.2f}h, gt={row['duration_hours_gt']:.2f}h, "
              f"diff={row['duration_diff_hours']:.2f}h")
    
    print(f"\nTop 10 combinations with largest duration differences (gt_df longer):")
    top_gt = in_both.nsmallest(10, 'duration_diff_hours')
    for idx, row in top_gt.iterrows():
        print(f"  ID {row['id']}, Obs {row['obs']}: "
              f"cam={row['duration_hours_cam']:.2f}h, gt={row['duration_hours_gt']:.2f}h, "
              f"diff={row['duration_diff_hours']:.2f}h")

# Combinations only in one dataset
only_cam = comparison[comparison['duration_hours_gt'].isna()]
only_gt = comparison[comparison['duration_hours_cam'].isna()]

print(f"\n{'='*60}")
print("COMBINATIONS IN ONLY ONE DATASET")
print(f"{'='*60}")
print(f"  Only in cam_df: {len(only_cam):,} combinations")
if len(only_cam) > 0:
    print(f"    Average duration: {only_cam['duration_hours_cam'].mean():.2f} hours")
    print(f"    Total duration: {only_cam['duration_hours_cam'].sum():.2f} hours")
print(f"  Only in gt_df: {len(only_gt):,} combinations")
if len(only_gt) > 0:
    print(f"    Average duration: {only_gt['duration_hours_gt'].mean():.2f} hours")
    print(f"    Total duration: {only_gt['duration_hours_gt'].sum():.2f} hours")

print(f"\n{'='*60}")
print("TIME COVERAGE ANALYSIS COMPLETE")
print(f"{'='*60}")
```

```python
# Standardize column names for comparison
# Identify ID and observation columns
gt1_id_col = id_cols_gt1[0] if id_cols_gt1 else None
gt1_obs_col = obs_cols_gt1[0] if obs_cols_gt1 else None
gt2_id_col = id_cols_gt2[0] if id_cols_gt2 else None
gt2_obs_col = obs_cols_gt2[0] if obs_cols_gt2 else None

print("="*60)
print("ID AND OBSERVATION COMBINATIONS COMPARISON")
print("="*60)

if gt1_id_col and gt1_obs_col and gt2_id_col and gt2_obs_col:
    # Get unique combinations
    gt1_combinations = groundtruth_df[[gt1_id_col, gt1_obs_col]].drop_duplicates()
    gt2_combinations = ground_gt_df[[gt2_id_col, gt2_obs_col]].drop_duplicates()
    
    # Standardize column names for comparison
    gt1_combinations_std = gt1_combinations.rename(columns={gt1_id_col: 'id', gt1_obs_col: 'obs'})
    gt2_combinations_std = gt2_combinations.rename(columns={gt2_id_col: 'id', gt2_obs_col: 'obs'})
    
    # Create sets for comparison
    gt1_combinations_set = set(zip(gt1_combinations_std['id'], gt1_combinations_std['obs']))
    gt2_combinations_set = set(zip(gt2_combinations_std['id'], gt2_combinations_std['obs']))
    
    # Find overlaps and differences
    common_combinations = gt1_combinations_set & gt2_combinations_set
    only_in_gt1 = gt1_combinations_set - gt2_combinations_set
    only_in_gt2 = gt2_combinations_set - gt1_combinations_set
    
    print(f"\nTotal unique (ID, Obs) combinations:")
    print(f"  groundtruth_df: {len(gt1_combinations_set):,}")
    print(f"  ground_gt_df: {len(gt2_combinations_set):,}")
    print(f"  In BOTH: {len(common_combinations):,}")
    print(f"  Only in groundtruth_df: {len(only_in_gt1):,}")
    print(f"  Only in ground_gt_df: {len(only_in_gt2):,}")
    
    # Count rows for each combination
    gt1_row_counts = groundtruth_df.groupby([gt1_id_col, gt1_obs_col]).size().reset_index(name='row_count')
    gt2_row_counts = ground_gt_df.groupby([gt2_id_col, gt2_obs_col]).size().reset_index(name='row_count')
    
    print(f"\n{'='*60}")
    print("ROW COUNTS BY COMBINATION")
    print(f"{'='*60}")
    print(f"\ngroundtruth_df:")
    print(f"  Total rows: {len(groundtruth_df):,}")
    print(f"  Average rows per combination: {gt1_row_counts['row_count'].mean():.1f}")
    print(f"  Median rows per combination: {gt1_row_counts['row_count'].median():.1f}")
    print(f"  Min rows: {gt1_row_counts['row_count'].min():,}")
    print(f"  Max rows: {gt1_row_counts['row_count'].max():,}")
    
    print(f"\nground_gt_df:")
    print(f"  Total rows: {len(ground_gt_df):,}")
    print(f"  Average rows per combination: {gt2_row_counts['row_count'].mean():.1f}")
    print(f"  Median rows per combination: {gt2_row_counts['row_count'].median():.1f}")
    print(f"  Min rows: {gt2_row_counts['row_count'].min():,}")
    print(f"  Max rows: {gt2_row_counts['row_count'].max():,}")
    
    # Show combinations only in groundtruth_df
    if len(only_in_gt1) > 0:
        print(f"\n{'='*60}")
        print(f"COMBINATIONS ONLY IN groundtruth_df ({len(only_in_gt1):,})")
        print(f"{'='*60}")
        only_gt1_list = list(only_in_gt1)[:20]  # Show first 20
        for id_val, obs_val in only_gt1_list:
            rows = len(groundtruth_df[(groundtruth_df[gt1_id_col] == id_val) & 
                                      (groundtruth_df[gt1_obs_col] == obs_val)])
            print(f"  ID {id_val}, Obs {obs_val}: {rows:,} rows")
        if len(only_in_gt1) > 20:
            print(f"  ... and {len(only_in_gt1) - 20} more")
    
    # Show combinations only in ground_gt_df
    if len(only_in_gt2) > 0:
        print(f"\n{'='*60}")
        print(f"COMBINATIONS ONLY IN ground_gt_df ({len(only_in_gt2):,})")
        print(f"{'='*60}")
        only_gt2_list = list(only_in_gt2)[:20]  # Show first 20
        for id_val, obs_val in only_gt2_list:
            rows = len(ground_gt_df[(ground_gt_df[gt2_id_col] == id_val) & 
                                    (ground_gt_df[gt2_obs_col] == obs_val)])
            print(f"  ID {id_val}, Obs {obs_val}: {rows:,} rows")
        if len(only_in_gt2) > 20:
            print(f"  ... and {len(only_in_gt2) - 20} more")
    
    # Compare row counts for common combinations
    if len(common_combinations) > 0:
        print(f"\n{'='*60}")
        print("ROW COUNT COMPARISON FOR COMMON COMBINATIONS")
        print(f"{'='*60}")
        
        # Merge row counts
        gt1_merged = gt1_row_counts.rename(columns={gt1_id_col: 'id', gt1_obs_col: 'obs', 'row_count': 'rows_gt1'})
        gt2_merged = gt2_row_counts.rename(columns={gt2_id_col: 'id', gt2_obs_col: 'obs', 'row_count': 'rows_gt2'})
        comparison = gt1_merged.merge(gt2_merged, on=['id', 'obs'], how='inner')
        comparison['row_diff'] = comparison['rows_gt1'] - comparison['rows_gt2']
        
        print(f"\n  Total common combinations: {len(comparison):,}")
        print(f"  Average row difference (gt1 - gt2): {comparison['row_diff'].mean():.1f}")
        print(f"  Median row difference: {comparison['row_diff'].median():.1f}")
        print(f"  Combinations with more rows in groundtruth_df: {(comparison['row_diff'] > 0).sum()}")
        print(f"  Combinations with more rows in ground_gt_df: {(comparison['row_diff'] < 0).sum()}")
        print(f"  Combinations with same row count: {(comparison['row_diff'] == 0).sum()}")
        
        # Show top differences
        if len(comparison) > 0:
            print(f"\n  Top 10 combinations with largest differences (gt1 - gt2):")
            top_diff = comparison.nlargest(10, 'row_diff')[['id', 'obs', 'rows_gt1', 'rows_gt2', 'row_diff']]
            for idx, row in top_diff.iterrows():
                print(f"    ID {row['id']}, Obs {row['obs']}: gt1={row['rows_gt1']:,}, gt2={row['rows_gt2']:,}, diff={row['row_diff']:+,}")
else:
    print("ERROR: Could not identify ID and observation columns in one or both dataframes")
    print(f"groundtruth_df ID cols: {id_cols_gt1}, Obs cols: {obs_cols_gt1}")
    print(f"ground_gt_df ID cols: {id_cols_gt2}, Obs cols: {obs_cols_gt2}")
```

### Analysis 2: Time Coverage Comparison

```python
# Time coverage analysis
print("="*60)
print("TIME COVERAGE COMPARISON")
print("="*60)

# Helper function to convert time string to seconds
def time_to_seconds(time_str):
    if pd.isna(time_str):
        return None
    time_str = str(time_str).strip()
    parts = time_str.split(':')
    if len(parts) == 3:
        try:
            return int(parts[0]) * 3600 + int(parts[1]) * 60 + int(float(parts[2]))
        except:
            return None
    return None

if gt1_id_col and gt1_obs_col and gt2_id_col and gt2_obs_col and time_cols_gt1 and time_cols_gt2:
    # Identify time column
    gt1_time_col = time_cols_gt1[0]
    gt2_time_col = time_cols_gt2[0]
    
    # Convert time to seconds for analysis
    groundtruth_df['_time_seconds'] = groundtruth_df[gt1_time_col].apply(time_to_seconds)
    ground_gt_df['_time_seconds'] = ground_gt_df[gt2_time_col].apply(time_to_seconds)
    
    # Calculate time coverage for each (id, obs) combination
    gt1_coverage = groundtruth_df.groupby([gt1_id_col, gt1_obs_col]).agg({
        '_time_seconds': ['min', 'max', 'count'],
        gt1_time_col: ['first', 'last']
    }).reset_index()
    gt1_coverage.columns = ['id', 'obs', 'min_seconds', 'max_seconds', 'row_count', 'first_time', 'last_time']
    gt1_coverage['duration_seconds'] = gt1_coverage['max_seconds'] - gt1_coverage['min_seconds']
    gt1_coverage['duration_hours'] = gt1_coverage['duration_seconds'] / 3600
    
    gt2_coverage = ground_gt_df.groupby([gt2_id_col, gt2_obs_col]).agg({
        '_time_seconds': ['min', 'max', 'count'],
        gt2_time_col: ['first', 'last']
    }).reset_index()
    gt2_coverage.columns = ['id', 'obs', 'min_seconds', 'max_seconds', 'row_count', 'first_time', 'last_time']
    gt2_coverage['duration_seconds'] = gt2_coverage['max_seconds'] - gt2_coverage['min_seconds']
    gt2_coverage['duration_hours'] = gt2_coverage['duration_seconds'] / 3600
    
    print(f"\ngroundtruth_df time coverage:")
    print(f"  Total combinations: {len(gt1_coverage):,}")
    print(f"  Average duration: {gt1_coverage['duration_hours'].mean():.2f} hours")
    print(f"  Median duration: {gt1_coverage['duration_hours'].median():.2f} hours")
    print(f"  Min duration: {gt1_coverage['duration_hours'].min():.2f} hours")
    print(f"  Max duration: {gt1_coverage['duration_hours'].max():.2f} hours")
    print(f"  Average row count: {gt1_coverage['row_count'].mean():.1f}")
    
    print(f"\nground_gt_df time coverage:")
    print(f"  Total combinations: {len(gt2_coverage):,}")
    print(f"  Average duration: {gt2_coverage['duration_hours'].mean():.2f} hours")
    print(f"  Median duration: {gt2_coverage['duration_hours'].median():.2f} hours")
    print(f"  Min duration: {gt2_coverage['duration_hours'].min():.2f} hours")
    print(f"  Max duration: {gt2_coverage['duration_hours'].max():.2f} hours")
    print(f"  Average row count: {gt2_coverage['row_count'].mean():.1f}")
    
    # Compare for common combinations
    comparison_time = gt1_coverage.merge(gt2_coverage, on=['id', 'obs'], how='inner', suffixes=('_gt1', '_gt2'))
    if len(comparison_time) > 0:
        comparison_time['duration_diff_hours'] = comparison_time['duration_hours_gt1'] - comparison_time['duration_hours_gt2']
        comparison_time['row_count_diff'] = comparison_time['row_count_gt1'] - comparison_time['row_count_gt2']
        
        print(f"\n{'='*60}")
        print("TIME COVERAGE COMPARISON FOR COMMON COMBINATIONS")
        print(f"{'='*60}")
        print(f"  Common combinations: {len(comparison_time):,}")
        print(f"  Average duration difference (gt1 - gt2): {comparison_time['duration_diff_hours'].mean():.3f} hours")
        print(f"  Median duration difference: {comparison_time['duration_diff_hours'].median():.3f} hours")
        print(f"  Average row count difference (gt1 - gt2): {comparison_time['row_count_diff'].mean():.1f}")
        print(f"  Median row count difference: {comparison_time['row_count_diff'].median():.1f}")
        
        # Show top differences
        print(f"\n  Top 10 combinations with largest duration differences (gt1 - gt2):")
        top_duration_diff = comparison_time.nlargest(10, 'duration_diff_hours')[['id', 'obs', 'duration_hours_gt1', 'duration_hours_gt2', 'duration_diff_hours']]
        for idx, row in top_duration_diff.iterrows():
            print(f"    ID {row['id']}, Obs {row['obs']}: gt1={row['duration_hours_gt1']:.2f}h, gt2={row['duration_hours_gt2']:.2f}h, diff={row['duration_diff_hours']:+.3f}h")
        
        # Visualizations
        import matplotlib.pyplot as plt
        
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        
        # Duration histograms
        axes[0, 0].hist(gt1_coverage['duration_hours'], bins=30, alpha=0.7, label='groundtruth_df', color='steelblue')
        axes[0, 0].hist(gt2_coverage['duration_hours'], bins=30, alpha=0.7, label='ground_gt_df', color='coral')
        axes[0, 0].set_xlabel('Duration (hours)')
        axes[0, 0].set_ylabel('Frequency')
        axes[0, 0].set_title('Duration Distribution by (ID, Obs) Combination')
        axes[0, 0].legend()
        axes[0, 0].grid(alpha=0.3)
        
        # Duration scatter plot
        axes[0, 1].scatter(comparison_time['duration_hours_gt1'], comparison_time['duration_hours_gt2'], alpha=0.5)
        axes[0, 1].plot([0, comparison_time['duration_hours_gt1'].max()], 
                        [0, comparison_time['duration_hours_gt1'].max()], 'r--', label='y=x')
        axes[0, 1].set_xlabel('groundtruth_df Duration (hours)')
        axes[0, 1].set_ylabel('ground_gt_df Duration (hours)')
        axes[0, 1].set_title('Duration Comparison: Common Combinations')
        axes[0, 1].legend()
        axes[0, 1].grid(alpha=0.3)
        
        # Row count scatter plot
        axes[1, 0].scatter(comparison_time['row_count_gt1'], comparison_time['row_count_gt2'], alpha=0.5)
        max_rows = max(comparison_time['row_count_gt1'].max(), comparison_time['row_count_gt2'].max())
        axes[1, 0].plot([0, max_rows], [0, max_rows], 'r--', label='y=x')
        axes[1, 0].set_xlabel('groundtruth_df Row Count')
        axes[1, 0].set_ylabel('ground_gt_df Row Count')
        axes[1, 0].set_title('Row Count Comparison: Common Combinations')
        axes[1, 0].legend()
        axes[1, 0].grid(alpha=0.3)
        
        # Duration difference histogram
        axes[1, 1].hist(comparison_time['duration_diff_hours'], bins=30, alpha=0.7, color='green')
        axes[1, 1].axvline(0, color='red', linestyle='--', linewidth=2, label='No difference')
        axes[1, 1].set_xlabel('Duration Difference (hours, gt1 - gt2)')
        axes[1, 1].set_ylabel('Frequency')
        axes[1, 1].set_title('Distribution of Duration Differences')
        axes[1, 1].legend()
        axes[1, 1].grid(alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Clean up temporary columns
        groundtruth_df.drop(columns=['_time_seconds'], inplace=True, errors='ignore')
        ground_gt_df.drop(columns=['_time_seconds'], inplace=True, errors='ignore')
else:
    print("ERROR: Could not identify required columns for time analysis")
    print(f"ID/Obs columns identified: gt1={gt1_id_col}/{gt1_obs_col}, gt2={gt2_id_col}/{gt2_obs_col}")
    print(f"Time columns: gt1={time_cols_gt1}, gt2={time_cols_gt2}")
```

### Analysis 3: Step Data Comparison

```python
# Step data comparison
print("="*60)
print("STEP DATA COMPARISON")
print("="*60)

if step_cols_gt1 and step_cols_gt2 and gt1_id_col and gt1_obs_col and gt2_id_col and gt2_obs_col:
    gt1_step_col = step_cols_gt1[0]
    gt2_step_col = step_cols_gt2[0]
    
    # Basic statistics
    print(f"\ngroundtruth_df step statistics:")
    print(f"  Total rows: {len(groundtruth_df):,}")
    print(f"  Rows with step data: {groundtruth_df[gt1_step_col].notna().sum():,}")
    print(f"  Rows with missing step: {groundtruth_df[gt1_step_col].isna().sum():,}")
    if groundtruth_df[gt1_step_col].notna().any():
        print(f"  Min step value: {groundtruth_df[gt1_step_col].min()}")
        print(f"  Max step value: {groundtruth_df[gt1_step_col].max()}")
        print(f"  Mean step value: {groundtruth_df[gt1_step_col].mean():.2f}")
        print(f"  Median step value: {groundtruth_df[gt1_step_col].median():.2f}")
        print(f"  Unique step values: {groundtruth_df[gt1_step_col].nunique():,}")
    
    print(f"\nground_gt_df step statistics:")
    print(f"  Total rows: {len(ground_gt_df):,}")
    print(f"  Rows with step data: {ground_gt_df[gt2_step_col].notna().sum():,}")
    print(f"  Rows with missing step: {ground_gt_df[gt2_step_col].isna().sum():,}")
    if ground_gt_df[gt2_step_col].notna().any():
        print(f"  Min step value: {ground_gt_df[gt2_step_col].min()}")
        print(f"  Max step value: {ground_gt_df[gt2_step_col].max()}")
        print(f"  Mean step value: {ground_gt_df[gt2_step_col].mean():.2f}")
        print(f"  Median step value: {ground_gt_df[gt2_step_col].median():.2f}")
        print(f"  Unique step values: {ground_gt_df[gt2_step_col].nunique():,}")
    
    # Compare unique step values
    if groundtruth_df[gt1_step_col].notna().any() and ground_gt_df[gt2_step_col].notna().any():
        gt1_unique_steps = set(groundtruth_df[gt1_step_col].dropna().unique())
        gt2_unique_steps = set(ground_gt_df[gt2_step_col].dropna().unique())
        
        common_steps = gt1_unique_steps & gt2_unique_steps
        only_gt1_steps = gt1_unique_steps - gt2_unique_steps
        only_gt2_steps = gt2_unique_steps - gt1_unique_steps
        
        print(f"\n{'='*60}")
        print("UNIQUE STEP VALUES COMPARISON")
        print(f"{'='*60}")
        print(f"  Unique step values in groundtruth_df: {len(gt1_unique_steps):,}")
        print(f"  Unique step values in ground_gt_df: {len(gt2_unique_steps):,}")
        print(f"  Common step values: {len(common_steps):,}")
        print(f"  Only in groundtruth_df: {len(only_gt1_steps):,}")
        print(f"  Only in ground_gt_df: {len(only_gt2_steps):,}")
        
        if len(only_gt1_steps) > 0 and len(only_gt1_steps) <= 20:
            print(f"\n  Step values only in groundtruth_df: {sorted(list(only_gt1_steps))}")
        elif len(only_gt1_steps) > 20:
            print(f"\n  Step values only in groundtruth_df (first 20): {sorted(list(only_gt1_steps))[:20]}")
        
        if len(only_gt2_steps) > 0 and len(only_gt2_steps) <= 20:
            print(f"\n  Step values only in ground_gt_df: {sorted(list(only_gt2_steps))}")
        elif len(only_gt2_steps) > 20:
            print(f"\n  Step values only in ground_gt_df (first 20): {sorted(list(only_gt2_steps))[:20]}")
    
    # Direct step comparison for matching rows (if time columns are available)
    if time_cols_gt1 and time_cols_gt2:
        gt1_time_col = time_cols_gt1[0]
        gt2_time_col = time_cols_gt2[0]
        
        # Normalize time format for merging
        def normalize_time(time_str):
            if pd.isna(time_str):
                return None
            time_str = str(time_str).strip()
            parts = time_str.split(':')
            if len(parts) == 3:
                try:
                    seconds_part = str(int(float(parts[2])))
                    return f"{int(parts[0]):02d}:{int(parts[1]):02d}:{seconds_part.zfill(2)}"
                except:
                    return time_str
            return time_str
        
        # Create merge keys
        groundtruth_df['_time_norm'] = groundtruth_df[gt1_time_col].apply(normalize_time)
        ground_gt_df['_time_norm'] = ground_gt_df[gt2_time_col].apply(normalize_time)
        
        # Merge on id, obs, and time
        merged_steps = groundtruth_df[[gt1_id_col, gt1_obs_col, '_time_norm', gt1_step_col]].rename(
            columns={gt1_id_col: 'id', gt1_obs_col: 'obs', gt1_step_col: 'step_gt1'}
        ).merge(
            ground_gt_df[[gt2_id_col, gt2_obs_col, '_time_norm', gt2_step_col]].rename(
                columns={gt2_id_col: 'id', gt2_obs_col: 'obs', gt2_step_col: 'step_gt2'}
            ),
            on=['id', 'obs', '_time_norm'],
            how='inner'
        )
        
        if len(merged_steps) > 0:
            print(f"\n{'='*60}")
            print("DIRECT STEP VALUE COMPARISON (Matching Rows)")
            print(f"{'='*60}")
            print(f"  Matching rows (same id, obs, time): {len(merged_steps):,}")
            
            # Compare step values
            both_have_step = merged_steps['step_gt1'].notna() & merged_steps['step_gt2'].notna()
            matches = (merged_steps.loc[both_have_step, 'step_gt1'] == merged_steps.loc[both_have_step, 'step_gt2']).sum()
            mismatches = both_have_step.sum() - matches
            
            print(f"  Rows where both have step data: {both_have_step.sum():,}")
            print(f"  Step values match: {matches:,} ({matches/both_have_step.sum()*100:.1f}%)")
            print(f"  Step values differ: {mismatches:,} ({mismatches/both_have_step.sum()*100:.1f}%)")
            
            if mismatches > 0:
                # Calculate differences for numeric steps
                numeric_mask = pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt1'], errors='coerce').notna() & \
                              pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt2'], errors='coerce').notna()
                if numeric_mask.any():
                    step1_numeric = pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt1'], errors='coerce')
                    step2_numeric = pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt2'], errors='coerce')
                    step_diff = step1_numeric - step2_numeric
                    
                    print(f"\n  Step difference statistics (where both are numeric):")
                    print(f"    Mean difference (gt1 - gt2): {step_diff[numeric_mask].mean():.2f}")
                    print(f"    Median difference: {step_diff[numeric_mask].median():.2f}")
                    print(f"    Min difference: {step_diff[numeric_mask].min():.2f}")
                    print(f"    Max difference: {step_diff[numeric_mask].max():.2f}")
                    print(f"    Std deviation: {step_diff[numeric_mask].std():.2f}")
            
            # Visualizations
            import matplotlib.pyplot as plt
            
            fig, axes = plt.subplots(2, 2, figsize=(16, 12))
            
            # Step distribution histograms
            if merged_steps['step_gt1'].notna().any():
                axes[0, 0].hist(merged_steps['step_gt1'].dropna(), bins=50, alpha=0.7, label='groundtruth_df', color='steelblue')
            if merged_steps['step_gt2'].notna().any():
                axes[0, 0].hist(merged_steps['step_gt2'].dropna(), bins=50, alpha=0.7, label='ground_gt_df', color='coral')
            axes[0, 0].set_xlabel('Step Value')
            axes[0, 0].set_ylabel('Frequency')
            axes[0, 0].set_title('Step Value Distribution (Matching Rows)')
            axes[0, 0].legend()
            axes[0, 0].grid(alpha=0.3)
            
            # Missing vs non-missing
            gt1_missing = merged_steps['step_gt1'].isna().sum()
            gt1_not_missing = merged_steps['step_gt1'].notna().sum()
            gt2_missing = merged_steps['step_gt2'].isna().sum()
            gt2_not_missing = merged_steps['step_gt2'].notna().sum()
            
            axes[0, 1].bar(['Missing', 'Not Missing'], [gt1_missing, gt1_not_missing], alpha=0.7, label='groundtruth_df', color='steelblue')
            axes[0, 1].bar(['Missing', 'Not Missing'], [gt2_missing, gt2_not_missing], alpha=0.7, label='ground_gt_df', color='coral', width=0.5)
            axes[0, 1].set_ylabel('Count')
            axes[0, 1].set_title('Missing vs Non-Missing Step Values')
            axes[0, 1].legend()
            axes[0, 1].grid(alpha=0.3, axis='y')
            
            # Step value scatter plot (where both have values)
            if both_have_step.sum() > 0:
                numeric_both = pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt1'], errors='coerce').notna() & \
                              pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt2'], errors='coerce').notna()
                if numeric_both.any():
                    step1_plot = pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt1'], errors='coerce')
                    step2_plot = pd.to_numeric(merged_steps.loc[both_have_step, 'step_gt2'], errors='coerce')
                    axes[1, 0].scatter(step1_plot[numeric_both], step2_plot[numeric_both], alpha=0.3)
                    max_step = max(step1_plot[numeric_both].max(), step2_plot[numeric_both].max())
                    axes[1, 0].plot([0, max_step], [0, max_step], 'r--', label='y=x')
                    axes[1, 0].set_xlabel('groundtruth_df Step Value')
                    axes[1, 0].set_ylabel('ground_gt_df Step Value')
                    axes[1, 0].set_title('Step Value Comparison (Numeric Values)')
                    axes[1, 0].legend()
                    axes[1, 0].grid(alpha=0.3)
            
            # Step difference histogram
            if mismatches > 0 and numeric_mask.any():
                axes[1, 1].hist(step_diff[numeric_mask], bins=50, alpha=0.7, color='green')
                axes[1, 1].axvline(0, color='red', linestyle='--', linewidth=2, label='No difference')
                axes[1, 1].set_xlabel('Step Difference (gt1 - gt2)')
                axes[1, 1].set_ylabel('Frequency')
                axes[1, 1].set_title('Distribution of Step Differences')
                axes[1, 1].legend()
                axes[1, 1].grid(alpha=0.3)
            
            plt.tight_layout()
            plt.show()
            
            # Clean up temporary columns
            groundtruth_df.drop(columns=['_time_norm'], inplace=True, errors='ignore')
            ground_gt_df.drop(columns=['_time_norm'], inplace=True, errors='ignore')
else:
    print("ERROR: Could not identify step columns or ID/Obs columns")
    print(f"Step columns: gt1={step_cols_gt1}, gt2={step_cols_gt2}")
    print(f"ID/Obs columns: gt1={gt1_id_col}/{gt1_obs_col}, gt2={gt2_id_col}/{gt2_obs_col}")
```
