## Step 2: Visualization Setup and Functions

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 6)
plt.rcParams['font.size'] = 10

print("Visualization libraries imported successfully!")
print("\nPackages being used:")
print("- matplotlib: For creating pie charts and bar charts")
print("- seaborn: For enhanced styling and color palettes")
print("- pandas: For data manipulation (already imported)")
```

```python
# Helper function to create comparison visualizations
def compare_categorical_columns(cam_col, gt_col, cam_df, gt_df, 
                                max_categories_for_pie=8, figsize=(16, 6)):
    """
    Compare categorical columns from two dataframes.
    Uses pie charts for small categories, bar charts for many categories.
    """
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # Get value counts for both columns
    cam_counts = cam_df[cam_col].value_counts()
    gt_counts = gt_df[gt_col].value_counts()
    
    # Determine if we should use pie or bar chart
    use_pie = len(cam_counts) <= max_categories_for_pie and len(gt_counts) <= max_categories_for_pie
    
    if use_pie:
        # Pie charts
        # cam_df pie chart
        axes[0].pie(cam_counts.values, labels=cam_counts.index, autopct='%1.1f%%', startangle=90)
        axes[0].set_title(f'cam_df: {cam_col}\n(n={len(cam_df)} rows)', fontsize=12, fontweight='bold')
        
        # gt_df pie chart
        axes[1].pie(gt_counts.values, labels=gt_counts.index, autopct='%1.1f%%', startangle=90)
        axes[1].set_title(f'gt_df: {gt_col}\n(n={len(gt_df)} rows)', fontsize=12, fontweight='bold')
    else:
        # Bar charts for many categories
        # cam_df bar chart
        cam_counts.plot(kind='bar', ax=axes[0], color='steelblue')
        axes[0].set_title(f'cam_df: {cam_col}\n(n={len(cam_df)} rows)', fontsize=12, fontweight='bold')
        axes[0].set_xlabel('Category')
        axes[0].set_ylabel('Count')
        axes[0].tick_params(axis='x', rotation=45)
        plt.setp(axes[0].xaxis.get_majorticklabels(), ha='right')
        axes[0].grid(axis='y', alpha=0.3)
        
        # gt_df bar chart
        gt_counts.plot(kind='bar', ax=axes[1], color='coral')
        axes[1].set_title(f'gt_df: {gt_col}\n(n={len(gt_df)} rows)', fontsize=12, fontweight='bold')
        axes[1].set_xlabel('Category')
        axes[1].set_ylabel('Count')
        axes[1].tick_params(axis='x', rotation=45)
        plt.setp(axes[1].xaxis.get_majorticklabels(), ha='right')
        axes[1].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print summary statistics
    print(f"\n{'='*60}")
    print(f"SUMMARY: {cam_col} (cam_df) vs {gt_col} (gt_df)")
    print(f"{'='*60}")
    print(f"\ncam_df - {cam_col}:")
    print(f"  Total unique values: {cam_df[cam_col].nunique()}")
    print(f"  Missing values: {cam_df[cam_col].isnull().sum()}")
    print(f"  Top 5 values:")
    for val, count in cam_counts.head(5).items():
        pct = (count / len(cam_df)) * 100
        print(f"    {val}: {count:,} ({pct:.1f}%)")
    
    print(f"\ngt_df - {gt_col}:")
    print(f"  Total unique values: {gt_df[gt_col].nunique()}")
    print(f"  Missing values: {gt_df[gt_col].isnull().sum()}")
    print(f"  Top 5 values:")
    for val, count in gt_counts.head(5).items():
        pct = (count / len(gt_df)) * 100
        print(f"    {val}: {count:,} ({pct:.1f}%)")
    print()
```

## Analysis: ID and Observation Combinations

```python
# Find total ID and obs combinations for both dataframes
print("="*60)
print("ID AND OBSERVATION COMBINATIONS ANALYSIS")
print("="*60)

# Get unique (id, obs) combinations for each dataframe
cam_combinations = cam_df[['id', 'obs']].drop_duplicates()
gt_combinations = gt_df[['id', 'observation']].drop_duplicates()

# Rename 'observation' to 'obs' in gt_df for consistency
gt_combinations_renamed = gt_combinations.rename(columns={'observation': 'obs'})

print(f"\ncam_df:")
print(f"  Total unique (ID, Obs) combinations: {len(cam_combinations):,}")
print(f"  Total rows: {len(cam_df):,}")

print(f"\ngt_df:")
print(f"  Total unique (ID, Obs) combinations: {len(gt_combinations_renamed):,}")
print(f"  Total rows: {len(gt_df):,}")

# Find combinations in cam_df but not in gt_df
cam_combinations_set = set(zip(cam_combinations['id'], cam_combinations['obs']))
gt_combinations_set = set(zip(gt_combinations_renamed['id'], gt_combinations_renamed['obs']))

only_in_cam = cam_combinations_set - gt_combinations_set
only_in_gt = gt_combinations_set - cam_combinations_set
in_both = cam_combinations_set & gt_combinations_set

print(f"\n{'='*60}")
print("COMBINATION OVERLAP")
print(f"{'='*60}")
print(f"  Combinations in BOTH dataframes: {len(in_both):,}")
print(f"  Combinations ONLY in cam_df: {len(only_in_cam):,}")
print(f"  Combinations ONLY in gt_df: {len(only_in_gt):,}")

# Count rows for combinations that are in cam_df but not in gt_df
only_in_cam_list = list(only_in_cam)
if len(only_in_cam_list) > 0:
    # Create a mask for rows in cam_df that match these combinations
    cam_mask = cam_df.apply(lambda row: (row['id'], row['obs']) in only_in_cam, axis=1)
    rows_only_in_cam = cam_df[cam_mask]
    
    print(f"\n{'='*60}")
    print("ROWS FOR COMBINATIONS IN cam_df BUT NOT IN gt_df")
    print(f"{'='*60}")
    print(f"  Number of rows: {len(rows_only_in_cam):,}")
    print(f"  Percentage of total cam_df: {(len(rows_only_in_cam)/len(cam_df)*100):.2f}%")
    
    # Show breakdown by combination
    combination_row_counts = rows_only_in_cam.groupby(['id', 'obs']).size().reset_index(name='row_count')
    print(f"\n  Breakdown by (ID, Obs) combination:")
    print(f"    Total unique combinations: {len(combination_row_counts):,}")
    print(f"    Average rows per combination: {combination_row_counts['row_count'].mean():.1f}")
    print(f"    Min rows per combination: {combination_row_counts['row_count'].min()}")
    print(f"    Max rows per combination: {combination_row_counts['row_count'].max()}")
    
    print(f"\n  Top 10 (ID, Obs) combinations by row count:")
    top_combos = combination_row_counts.nlargest(10, 'row_count')
    for idx, row in top_combos.iterrows():
        print(f"    ID {row['id']}, Obs {row['obs']}: {row['row_count']:,} rows")
else:
    print(f"\n  No combinations found only in cam_df")
```

## Analysis: Time Coverage Comparison by (ID, Obs) Combinations

```python
# Analyze time coverage for each unique (id, obs) combination
print("="*60)
print("TIME COVERAGE ANALYSIS BY (ID, OBS) COMBINATIONS")
print("="*60)

# Helper function to convert rel_time to seconds for calculations
def time_to_seconds(time_str):
    """Convert HH:MM:SS format to total seconds"""
    if pd.isna(time_str):
        return np.nan
    try:
        parts = str(time_str).split(':')
        if len(parts) == 3:
            hours = int(parts[0])
            minutes = int(parts[1])
            seconds = int(parts[2])
            return hours * 3600 + minutes * 60 + seconds
        return np.nan
    except:
        return np.nan

# Calculate time coverage for cam_df
cam_df['rel_time_seconds'] = cam_df['rel_time'].apply(time_to_seconds)
cam_coverage = cam_df.groupby(['id', 'obs']).agg({
    'rel_time_seconds': ['min', 'max', 'count'],
    'rel_time': ['first', 'last']
}).reset_index()
cam_coverage.columns = ['id', 'obs', 'min_seconds', 'max_seconds', 'row_count', 'first_rel_time', 'last_rel_time']
cam_coverage['duration_seconds'] = cam_coverage['max_seconds'] - cam_coverage['min_seconds']
cam_coverage['duration_hours'] = cam_coverage['duration_seconds'] / 3600

# Calculate time coverage for gt_df
gt_df['rel_time_seconds'] = gt_df['relative_time'].apply(time_to_seconds)
gt_coverage = gt_df.groupby(['id', 'observation']).agg({
    'rel_time_seconds': ['min', 'max', 'count'],
    'relative_time': ['first', 'last']
}).reset_index()
gt_coverage.columns = ['id', 'obs', 'min_seconds', 'max_seconds', 'row_count', 'first_rel_time', 'last_rel_time']
gt_coverage['duration_seconds'] = gt_coverage['max_seconds'] - gt_coverage['min_seconds']
gt_coverage['duration_hours'] = gt_coverage['duration_seconds'] / 3600

print(f"\ncam_df:")
print(f"  Total unique (ID, Obs) combinations: {len(cam_coverage):,}")
print(f"  Average duration per combination: {cam_coverage['duration_hours'].mean():.2f} hours")
print(f"  Median duration per combination: {cam_coverage['duration_hours'].median():.2f} hours")
print(f"  Min duration: {cam_coverage['duration_hours'].min():.2f} hours")
print(f"  Max duration: {cam_coverage['duration_hours'].max():.2f} hours")
print(f"  Average row count per combination: {cam_coverage['row_count'].mean():.1f}")

print(f"\ngt_df:")
print(f"  Total unique (ID, Obs) combinations: {len(gt_coverage):,}")
print(f"  Average duration per combination: {gt_coverage['duration_hours'].mean():.2f} hours")
print(f"  Median duration per combination: {gt_coverage['duration_hours'].median():.2f} hours")
print(f"  Min duration: {gt_coverage['duration_hours'].min():.2f} hours")
print(f"  Max duration: {gt_coverage['duration_hours'].max():.2f} hours")
print(f"  Average row count per combination: {gt_coverage['row_count'].mean():.1f}")

# Merge to compare coverage for matching (id, obs) combinations
comparison = cam_coverage.merge(
    gt_coverage,
    on=['id', 'obs'],
    how='outer',
    suffixes=('_cam', '_gt')
)

# Calculate differences
comparison['duration_diff_hours'] = comparison['duration_hours_cam'] - comparison['duration_hours_gt']
comparison['row_count_diff'] = comparison['row_count_cam'] - comparison['row_count_gt']

# Find combinations in both datasets
in_both = comparison[comparison['duration_hours_cam'].notna() & comparison['duration_hours_gt'].notna()]

print(f"\n{'='*60}")
print("COVERAGE COMPARISON FOR MATCHING (ID, OBS) COMBINATIONS")
print(f"{'='*60}")
print(f"  Combinations in BOTH datasets: {len(in_both):,}")
if len(in_both) > 0:
    print(f"  Average duration difference (cam - gt): {in_both['duration_diff_hours'].mean():.2f} hours")
    print(f"  Median duration difference: {in_both['duration_diff_hours'].median():.2f} hours")
    print(f"  Average row count difference (cam - gt): {in_both['row_count_diff'].mean():.1f} rows")
    print(f"  Median row count difference: {in_both['row_count_diff'].median():.1f} rows")
    
    # Count how many have longer/shorter coverage
    longer_in_cam = (in_both['duration_diff_hours'] > 0).sum()
    longer_in_gt = (in_both['duration_diff_hours'] < 0).sum()
    same_duration = (in_both['duration_diff_hours'] == 0).sum()
    
    print(f"\n  Duration comparison:")
    print(f"    Longer in cam_df: {longer_in_cam:,} ({longer_in_cam/len(in_both)*100:.1f}%)")
    print(f"    Longer in gt_df: {longer_in_gt:,} ({longer_in_gt/len(in_both)*100:.1f}%)")
    print(f"    Same duration: {same_duration:,} ({same_duration/len(in_both)*100:.1f}%)")

# Create visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Plot 1: Distribution of durations - cam_df
axes[0, 0].hist(cam_coverage['duration_hours'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Duration (hours)', fontsize=11)
axes[0, 0].set_ylabel('Number of (ID, Obs) Combinations', fontsize=11)
axes[0, 0].set_title(f'cam_df: Duration Distribution\n(n={len(cam_coverage):,} combinations)', 
                    fontsize=12, fontweight='bold')
axes[0, 0].grid(axis='y', alpha=0.3)

# Plot 2: Distribution of durations - gt_df
axes[0, 1].hist(gt_coverage['duration_hours'], bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[0, 1].set_xlabel('Duration (hours)', fontsize=11)
axes[0, 1].set_ylabel('Number of (ID, Obs) Combinations', fontsize=11)
axes[0, 1].set_title(f'gt_df: Duration Distribution\n(n={len(gt_coverage):,} combinations)', 
                    fontsize=12, fontweight='bold')
axes[0, 1].grid(axis='y', alpha=0.3)

# Plot 3: Side-by-side comparison of durations
if len(in_both) > 0:
    axes[0, 2].scatter(in_both['duration_hours_gt'], in_both['duration_hours_cam'], 
                       alpha=0.5, s=20, color='purple')
    # Add diagonal line
    max_val = max(in_both['duration_hours_gt'].max(), in_both['duration_hours_cam'].max())
    axes[0, 2].plot([0, max_val], [0, max_val], 'r--', linewidth=2, label='Equal duration')
    axes[0, 2].set_xlabel('gt_df Duration (hours)', fontsize=11)
    axes[0, 2].set_ylabel('cam_df Duration (hours)', fontsize=11)
    axes[0, 2].set_title(f'Duration Comparison: cam_df vs gt_df\n(n={len(in_both):,} matching combinations)', 
                        fontsize=12, fontweight='bold')
    axes[0, 2].legend()
    axes[0, 2].grid(alpha=0.3)

# Plot 4: Distribution of duration differences
if len(in_both) > 0:
    axes[1, 0].hist(in_both['duration_diff_hours'], bins=50, color='green', edgecolor='black', alpha=0.7)
    axes[1, 0].axvline(x=0, color='red', linestyle='--', linewidth=2, label='No difference')
    axes[1, 0].set_xlabel('Duration Difference (cam_df - gt_df) in hours', fontsize=11)
    axes[1, 0].set_ylabel('Number of Combinations', fontsize=11)
    axes[1, 0].set_title('Distribution of Duration Differences', fontsize=12, fontweight='bold')
    axes[1, 0].legend()
    axes[1, 0].grid(axis='y', alpha=0.3)

# Plot 5: Row count comparison
if len(in_both) > 0:
    axes[1, 1].scatter(in_both['row_count_gt'], in_both['row_count_cam'], 
                      alpha=0.5, s=20, color='orange')
    max_rows = max(in_both['row_count_gt'].max(), in_both['row_count_cam'].max())
    axes[1, 1].plot([0, max_rows], [0, max_rows], 'r--', linewidth=2, label='Equal row count')
    axes[1, 1].set_xlabel('gt_df Row Count', fontsize=11)
    axes[1, 1].set_ylabel('cam_df Row Count', fontsize=11)
    axes[1, 1].set_title(f'Row Count Comparison\n(n={len(in_both):,} matching combinations)', 
                        fontsize=12, fontweight='bold')
    axes[1, 1].legend()
    axes[1, 1].grid(alpha=0.3)

# Plot 6: Top 20 combinations by duration difference
if len(in_both) > 0:
    top_diff = in_both.nlargest(20, 'duration_diff_hours')[['id', 'obs', 'duration_diff_hours']]
    top_diff['label'] = top_diff.apply(lambda x: f"ID {x['id']}, Obs {x['obs']}", axis=1)
    
    axes[1, 2].barh(range(len(top_diff)), top_diff['duration_diff_hours'].values, 
                   color='steelblue', alpha=0.7)
    axes[1, 2].set_yticks(range(len(top_diff)))
    axes[1, 2].set_yticklabels(top_diff['label'], fontsize=9)
    axes[1, 2].set_xlabel('Duration Difference (hours)', fontsize=11)
    axes[1, 2].set_title('Top 20 Combinations: Longer in cam_df', fontsize=12, fontweight='bold')
    axes[1, 2].axvline(x=0, color='red', linestyle='--', linewidth=1)
    axes[1, 2].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Print detailed statistics
print(f"\n{'='*60}")
print("DETAILED STATISTICS")
print(f"{'='*60}")

if len(in_both) > 0:
    print(f"\nTop 10 combinations with largest duration differences (cam_df longer):")
    top_cam = in_both.nlargest(10, 'duration_diff_hours')
    for idx, row in top_cam.iterrows():
        print(f"  ID {row['id']}, Obs {row['obs']}: "
              f"cam={row['duration_hours_cam']:.2f}h, gt={row['duration_hours_gt']:.2f}h, "
              f"diff={row['duration_diff_hours']:.2f}h")
    
    print(f"\nTop 10 combinations with largest duration differences (gt_df longer):")
    top_gt = in_both.nsmallest(10, 'duration_diff_hours')
    for idx, row in top_gt.iterrows():
        print(f"  ID {row['id']}, Obs {row['obs']}: "
              f"cam={row['duration_hours_cam']:.2f}h, gt={row['duration_hours_gt']:.2f}h, "
              f"diff={row['duration_diff_hours']:.2f}h")

# Combinations only in one dataset
only_cam = comparison[comparison['duration_hours_gt'].isna()]
only_gt = comparison[comparison['duration_hours_cam'].isna()]

print(f"\n{'='*60}")
print("COMBINATIONS IN ONLY ONE DATASET")
print(f"{'='*60}")
print(f"  Only in cam_df: {len(only_cam):,} combinations")
if len(only_cam) > 0:
    print(f"    Average duration: {only_cam['duration_hours_cam'].mean():.2f} hours")
    print(f"    Total duration: {only_cam['duration_hours_cam'].sum():.2f} hours")
print(f"  Only in gt_df: {len(only_gt):,} combinations")
if len(only_gt) > 0:
    print(f"    Average duration: {only_gt['duration_hours_gt'].mean():.2f} hours")
    print(f"    Total duration: {only_gt['duration_hours_gt'].sum():.2f} hours")

print(f"\n{'='*60}")
print("TIME COVERAGE ANALYSIS COMPLETE")
print(f"{'='*60}")
```
